## Setup Step 3: Initialize SparkSession & Temp SQL Views
We startup the Spark session and register local temporary views so we can execute Spark SQL queries directly in python.


```python
from pyspark.sql import SparkSession

spark = SparkSession.builder \
    .appName('GoogleColabPySparkETL') \
    .getOrCreate()

# Read dataframes to register temp tables for SQL execution
customers_raw = spark.read.csv('data/customers.csv', header=True, inferSchema=True)
sales_raw = spark.read.csv('data/sales.csv', header=True, inferSchema=True)
products_raw = spark.read.csv('data/products.csv', header=True, inferSchema=True)

customers_raw.createOrReplaceTempView('customers')
sales_raw.createOrReplaceTempView('sales')
products_raw.createOrReplaceTempView('products')

print('SparkSession configured and views registered!')
```

## Exercise 2: Cleaning Nulls & Invalid Data
**Objective**: Detect missing entries, drop complete nulls, fill defaults, filter negative values.


```python
# ----------------- SQL Approach -----------------
print('=== SQL Approach (Null Counts) ===')
spark.sql('''
SELECT
    SUM(CASE WHEN customer_id IS NULL THEN 1 ELSE 0 END) AS null_cust_id,
    SUM(CASE WHEN city IS NULL THEN 1 ELSE 0 END) AS null_city
FROM customers
''').show()

# ----------------- PySpark Approach -----------------
print('=== PySpark Approach (Null Counts & Cleaning) ===')
from pyspark.sql.functions import col, count, when

customers_raw.select(
    [count(when(col(c).isNull(), c)).alias(f'null_{c}') for c in ['customer_id', 'city']]
).show()

# Fill Nulls and drop missing emails
clean_customers_df = customers_raw.dropna(subset=['customer_id', 'email']) \
    .fillna({'city': 'Unknown', 'state': 'Unknown'})
print(f'Customers before: {customers_raw.count()} | Cleaned: {clean_customers_df.count()}')
```

## Exercise 3: Daily Sales Analysis
**Objective**: Calculate daily transactions, revenue, and average transaction size.


```python
# ----------------- SQL Approach -----------------
print('=== SQL: Top 5 Revenue Days ===')
spark.sql('''
SELECT
    sale_date,
    COUNT(*) AS total_orders,
    ROUND(SUM(total_amount), 2) AS daily_revenue
FROM sales
WHERE quantity > 0 AND total_amount > 0
GROUP BY sale_date
ORDER BY daily_revenue DESC
LIMIT 5
''').show()

# ----------------- PySpark Approach -----------------
print('=== PySpark: Daily Sales Aggregation ===')
from pyspark.sql.functions import sum, round, avg

daily_sales_df = sales_raw.filter((col('quantity') > 0) & (col('total_amount') > 0)) \
    .groupBy('sale_date') \
    .agg(
        count('sale_id').alias('total_orders'),
        round(sum('total_amount'), 2).alias('daily_revenue'),
        round(avg('total_amount'), 2).alias('avg_order_value')
    ) \
    .orderBy(col('daily_revenue').desc())
daily_sales_df.show(5)
```

## Exercise 4: City-Wise Revenue Breakdown
**Objective**: Join transaction tables and customer profiles to trace geographical performance.


```python
# ----------------- SQL Approach -----------------
print('=== SQL: Top 5 Cities by Revenue ===')
spark.sql('''
SELECT
    c.city,
    c.state,
    ROUND(SUM(s.total_amount), 2) AS total_revenue
FROM customers c
INNER JOIN sales s ON c.customer_id = s.customer_id
WHERE c.city IS NOT NULL
GROUP BY c.city, c.state
ORDER BY total_revenue DESC
LIMIT 5
''').show()

# ----------------- PySpark Approach -----------------
print('=== PySpark: City Revenue aggregation ===')
city_revenue_df = clean_customers_df.join(sales_raw, on='customer_id', how='inner') \
    .groupBy('city', 'state') \
    .agg(round(sum('total_amount'), 2).alias('total_revenue')) \
    .orderBy(col('total_revenue').desc())
city_revenue_df.show(5)
```

## Exercise 5: Repeat Customer Tracking
**Objective**: Identify customers ordering frequently (>= 2 orders).


```python
# ----------------- SQL Approach -----------------
print('=== SQL: Repeat Customers (>=2 orders) ===')
spark.sql('''
SELECT
    c.customer_id,
    c.first_name,
    c.last_name,
    COUNT(s.sale_id) AS order_count
FROM customers c
INNER JOIN sales s ON c.customer_id = s.customer_id
GROUP BY c.customer_id, c.first_name, c.last_name
HAVING COUNT(s.sale_id) >= 2
ORDER BY order_count DESC
''').show()

# ----------------- PySpark Approach -----------------
print('=== PySpark: Repeat Customers ===')
repeat_cust = clean_customers_df.join(sales_raw, on='customer_id', how='inner') \
    .groupBy('customer_id', 'first_name', 'last_name') \
    .agg(count('sale_id').alias('order_count')) \
    .filter(col('order_count') >= 2) \
    .orderBy(col('order_count').desc())
repeat_cust.show(5)
```

## Exercise 6: Highest Spending Customer in Each City
**Objective**: Use Window Partitioning to target top spend ranking.


```python
# ----------------- SQL Approach -----------------
print('=== SQL: Top spender per city ===')
spark.sql('''
SELECT * FROM (
    SELECT
        c.city,
        c.first_name,
        c.last_name,
        SUM(s.total_amount) AS total_spent,
        ROW_NUMBER() OVER (PARTITION BY c.city ORDER BY SUM(s.total_amount) DESC) as rank
    FROM customers c
    INNER JOIN sales s ON c.customer_id = s.customer_id
    WHERE c.city IS NOT NULL
    GROUP BY c.city, c.first_name, c.last_name
) WHERE rank = 1
''').show()

# ----------------- PySpark Approach -----------------
print('=== PySpark: Top spender per city ===')
from pyspark.sql.window import Window
from pyspark.sql.functions import row_number

cust_spend = clean_customers_df.join(sales_raw, on='customer_id', how='inner') \
    .groupBy('city', 'customer_id', 'first_name', 'last_name') \
    .agg(sum('total_amount').alias('total_spent'))

city_window = Window.partitionBy('city').orderBy(col('total_spent').desc())
top_spender_pyspark = cust_spend.withColumn('rank', row_number().over(city_window)) \
    .filter(col('rank') == 1) \
    .drop('rank') \
    .orderBy(col('total_spent').desc())

top_spender_pyspark.show(5)
```

## Exercise 7: Final Consolidated Reporting Table
**Objective**: Enrich, join, and load customer summary statistics.


```python
# ----------------- SQL Approach -----------------
print('=== SQL: Consolidated Dashboard View ===')
spark.sql('''
SELECT
    c.customer_id,
    CONCAT(c.first_name, \' \', c.last_name) AS full_name,
    c.city,
    COUNT(s.sale_id) AS order_count,
    ROUND(SUM(s.total_amount), 2) AS total_spend
FROM customers c
INNER JOIN sales s ON c.customer_id = s.customer_id
WHERE c.city IS NOT NULL
GROUP BY c.customer_id, c.first_name, c.last_name, c.city
ORDER BY total_spend DESC
''').show()

# ----------------- PySpark Approach -----------------
print('=== PySpark: Reporting Table ===')
from pyspark.sql.functions import concat, lit

final_report = clean_customers_df.join(sales_raw, on='customer_id', how='inner') \
    .groupBy('customer_id', 'first_name', 'last_name', 'city') \
    .agg(
        count('sale_id').alias('order_count'),
        round(sum('total_amount'), 2).alias('total_spend')
    ) \
    .withColumn('full_name', concat(col('first_name'), lit(' '), col('last_name'))) \
    .select('customer_id', 'full_name', 'city', 'order_count', 'total_spend') \
    .orderBy(col('total_spend').desc())

final_report.show(5)
```

## Exercise 8: Reusable production-grade ETL Pipeline Class
This wraps all stages (`Extract`, `Transform`, `Load`) into a single reusable object with logging.


```python
import logging
import sys

logging.basicConfig(level=logging.INFO, format='%(asctime)s [%(levelname)s] %(message)s')
logger = logging.getLogger('RetailETL')

class RetailETLPipeline:
    def __init__(self, customers_path, sales_path, products_path):
        self.customers_path = customers_path
        self.sales_path = sales_path
        self.products_path = products_path
        self.spark = None

    def extract(self):
        logger.info('Extracting raw datasets...')
        self.spark = SparkSession.builder.getOrCreate()
        c_df = self.spark.read.csv(self.customers_path, header=True, inferSchema=True)
        s_df = self.spark.read.csv(self.sales_path, header=True, inferSchema=True)
        p_df = self.spark.read.csv(self.products_path, header=True, inferSchema=True)
        return c_df, s_df, p_df

    def transform(self, c_df, s_df, p_df):
        logger.info('Transforming and cleaning data...')
        # Clean datasets
        clean_c = c_df.dropna(subset=['customer_id', 'email']).fillna({'city': 'Unknown', 'state': 'Unknown'})
        clean_s = s_df.dropna().filter((col('quantity') > 0) & (col('total_amount') > 0))

        # Consolidated report
        final_report = clean_c.join(clean_s, on='customer_id', how='inner') \
            .groupBy('customer_id', 'first_name', 'last_name', 'city', 'state') \
            .agg(
                count('sale_id').alias('order_count'),
                round(sum('total_amount'), 2).alias('total_spend'),
                round(avg('total_amount'), 2).alias('avg_order_value')
            ) \
            .withColumn('full_name', concat(col('first_name'), lit(' '), col('last_name'))) \
            .select('customer_id', 'full_name', 'city', 'state', 'order_count', 'total_spend', 'avg_order_value') \
            .orderBy(col('total_spend').desc())
        return final_report

    def load(self, final_report):
        logger.info('Loading final results to storage...')
        # Mock write to local file system (CSV format for display)
        final_report.write.mode('overwrite').csv('data/output_report', header=True)
        print('\n=== ETL Pipeline Executed Successfully! Output preview: ===')
        final_report.show(10)

    def run(self):
        try:
            c, s, p = self.extract()
            res = self.transform(c, s, p)
            self.load(res)
        except Exception as e:
            logger.error(f'Pipeline error: {e}')

# Instantiate and run the pipeline
pipeline = RetailETLPipeline('data/customers.csv', 'data/sales.csv', 'data/products.csv')
pipeline.run()
```
