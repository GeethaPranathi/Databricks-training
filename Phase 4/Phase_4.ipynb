```python
from pyspark.sql import SparkSession
from pyspark.sql.functions import col, count, sum, when, concat_ws, desc
import logging

# Configure logging
logging.basicConfig(level=logging.INFO, format="%(asctime)s [%(levelname)s] %(message)s")
logger = logging.getLogger("BusinessPipeline")

logger.info("Initializing Spark Session...")
spark = SparkSession.builder \
    .appName("Phase4-BusinessPipeline") \
    .getOrCreate()

# Paths are now just the local files we just created
customers_path = "customers.csv"
sales_path = "sales.csv"
output_path = "output_report"

logger.info("Reading raw data...")
customers_df = spark.read.csv(customers_path, header=True, inferSchema=True)
sales_df = spark.read.csv(sales_path, header=True, inferSchema=True)

logger.info("Cleaning Data...")
# Clean customers: drop null keys
clean_customers = customers_df.dropna(subset=["customer_id"])
clean_customers = clean_customers.dropDuplicates()

# Clean sales: drop null keys, drop duplicates, filter invalid values
clean_sales = sales_df.dropna(subset=["customer_id", "sale_id"])
clean_sales = clean_sales.dropDuplicates()
# Assuming total_amount and quantity should be positive
clean_sales = clean_sales.filter((col("total_amount") > 0) & (col("quantity") > 0))

logger.info("Executing Pipeline Tasks...")

# Task 1: Daily Sales
daily_sales = clean_sales.groupBy("sale_date") \
    .agg(sum("total_amount").alias("total_sales")) \
    .orderBy("sale_date")
logger.info("Task 1: Daily Sales (Preview)")
daily_sales.show(5)

# Task 2: City-wise Revenue
city_revenue = clean_sales.join(clean_customers, "customer_id", "inner") \
    .groupBy("city") \
    .agg(sum("total_amount").alias("total_revenue")) \
    .orderBy(desc("total_revenue"))
logger.info("Task 2: City-wise Revenue (Preview)")
city_revenue.show(5)

# Task 3: Top 5 Customers
top_5_customers = clean_sales.join(clean_customers, "customer_id", "inner") \
    .withColumn("customer_name", concat_ws(" ", col("first_name"), col("last_name"))) \
    .groupBy("customer_name") \
    .agg(sum("total_amount").alias("total_spend")) \
    .orderBy(desc("total_spend")) \
    .limit(5)
logger.info("Task 3: Top 5 Customers")
top_5_customers.show()

# Task 4: Repeat Customers
repeat_customers = clean_sales.groupBy("customer_id") \
    .agg(count("sale_id").alias("order_count")) \
    .filter(col("order_count") > 1) \
    .orderBy(desc("order_count"))
logger.info("Task 4: Repeat Customers (Preview)")
repeat_customers.show(5)

# Task 5 & 6: Customer Segmentation & Final Reporting Table
customer_stats = clean_sales.groupBy("customer_id") \
    .agg(
        sum("total_amount").alias("total_spend"),
        count("sale_id").alias("order_count")
    )

final_df = customer_stats.join(clean_customers, "customer_id", "inner") \
    .withColumn("customer_name", concat_ws(" ", col("first_name"), col("last_name"))) \
    .withColumn(
        "segment",
        when(col("total_spend") > 10000, "Gold")
        .when((col("total_spend") >= 5000) & (col("total_spend") <= 10000), "Silver")
        .otherwise("Bronze")
    ) \
    .select("customer_name", "city", "total_spend", "order_count", "segment") \
    .orderBy(desc("total_spend"))

logger.info("Task 5 & 6: Final Reporting Table (Preview)")
final_df.show(10)

# Task 7: Save Output
logger.info(f"Task 7: Saving output to {output_path}")
final_df.write.mode("overwrite").csv(output_path, header=True)

logger.info("Pipeline completed successfully!")
```
